# Two variables plots

Adapted from <https://matplotlib.org/tutorials/introductory/pyplot.html>.

```python
import matplotlib.pyplot as plt
import numpy as np
```

$$ y = 2 x $$

```python
x = np.linspace(0.0, 100.0, 10)
y = -2.0 * x + 1.0

plt.plot(x, y, 'k.')
```

```python
x = np.linspace(0.0, 100.0, 10)
y1 = 2.0 * x + 5.0
y2 = 1.0 * x - 5.0
y3 = -2.0 * x + 0.0

plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')
plt.plot(x, y3, 'g')
```

```python
x = np.linspace(0.0, 10.0, 100)
y1 = np.exp(x)
y2 = 1.0 / np.exp(x)
#y3 = x**x

plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')
#plt.plot(x, y3, 'g')
```
